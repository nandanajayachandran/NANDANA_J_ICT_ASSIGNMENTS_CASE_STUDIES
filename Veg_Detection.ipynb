```python
import cv2
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
```

```python
#Load the saved model
model = load_model("vegetable_classification_model.h5")
```

```python
#Open the video capture
cap = cv2.VideoCapture(0)

#Create a dictionary to map class
class_labels = {0: 'carrot', 1: 'tomato', 2: 'lettuce'}

while True:
    #Capture a frame from the webcam
    ret, frame = cap.read()

    #Check if the frame was captured successfully
    if not ret:
        print("Error: Unable to capture frame from the webcam.")
        break

    #Preprocess the frame for the model (adjust as needed)
    processed_frame = cv2.resize(frame, (224, 224))
    processed_frame = processed_frame / 255.0
    processed_frame = np.expand_dims(processed_frame, axis=0)

    #Making predictions
    predictions = model.predict(processed_frame)

    #Get the predicted class index
    predicted_class_index = np.argmax(predictions[0])

    #Map the index to the class label
    predicted_class_label = class_labels.get(predicted_class_index, 'Unknown')

    #Display the result on the frame
    cv2.putText(frame, f"Predicted: {predicted_class_label}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    #Display the frame
    cv2.imshow("Vegetable Classification", frame)

    #Break the loop if 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

#Release the video capture and close the OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
