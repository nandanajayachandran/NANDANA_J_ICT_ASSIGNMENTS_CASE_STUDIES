Imputation - Train dataset

```python
# Imputing the missing values in categorical columns with mode and numericalcolumn with median
df_train['Gender'].fillna(df_train['Gender'].mode()[0], inplace=True)
df_train['Married'].fillna(df_train['Married'].mode()[0], inplace=True)
df_train['Dependents'].fillna(df_train['Dependents'].mode()[0], inplace=True)
df_train['Self_Employed'].fillna(df_train['Self_Employed'].mode()[0], inplace=True)
df_train['LoanAmount'].fillna(df_train['LoanAmount'].median(), inplace=True)
df_train['Loan_Amount_Term'].fillna(df_train['Loan_Amount_Term'].mode()[0], inplace=True)
df_train['Credit_History'].fillna(df_train['Credit_History'].mode()[0], inplace=True)
```

Imputation - Test Dataset

```python
# Imputing the missing values in categorical columns with mode and numerical columns with median
df_test['Gender'].fillna(df_test['Gender'].mode()[0], inplace=True)
df_test['Dependents'].fillna(df_test['Dependents'].mode()[0], inplace=True)
df_test['Self_Employed'].fillna(df_test['Self_Employed'].mode()[0], inplace=True)
df_test['LoanAmount'].fillna(df_test['LoanAmount'].median(), inplace=True)
df_test['Loan_Amount_Term'].fillna(df_test['Loan_Amount_Term'].mode()[0], inplace=True)
df_test['Credit_History'].fillna(df_test['Credit_History'].mode()[0], inplace=True)
```

```python
#Checking whether the null values are filled or not
#Checking for null values in df_train
df_train.isna().sum()
```

```python
#Checking for null values in df_test
df_test.isna().sum()
```

**OUTLIER DETECTION AND HANDLING**

```python
#Boxplot for df_train
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_train)
plt.title('Boxplot for numerical columns')
plt.xticks(rotation=30)
plt.show()
```

```python
#Boxplot for df_test
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_test)
plt.title('Boxplot for numerical columns')
plt.xticks(rotation=30)
plt.show()
```

We can see outliers in both df_train and df_test

```python
#Handling outliers
numerical_columns = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
# Applying log transformation to handle outliers
for column in numerical_columns:
    df_train[column] = np.log1p(df_train[column])
for column in numerical_columns:
    df_test[column] = np.log1p(df_test[column])
```

**ENCODING**

```python
#ONE HOT ENCODING
categorical_columns = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']

# Applying one-hot encoding to the train dataset
df_train_encoded = pd.get_dummies(df_train, columns=categorical_columns, drop_first=True)

# Applying one-hot encoding to the test dataset
df_test_encoded = pd.get_dummies(df_test, columns=categorical_columns, drop_first=True)
```

**SCALING**

```python
from sklearn.preprocessing import StandardScaler
# the numerical columns to be standardized
numerical_columns = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
scaler = StandardScaler()

# Applying standardization to the train dataset
df_train_scaled = df_train.copy()
df_train_scaled[numerical_columns] = scaler.fit_transform(df_train_scaled[numerical_columns])

# Applying standardization to the test dataset
df_test_scaled = df_test.copy()
df_test_scaled[numerical_columns] = scaler.transform(df_test_scaled[numerical_columns])
```
